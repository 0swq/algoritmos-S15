```python
class nodo:
    def __init__(self, valor):
        self.izquierdo = None
        self.derecho = None
        self.valor = valor


def insertar(raiz, valor):
    if raiz is None:
        return nodo(valor)
    if valor < raiz.valor:
        raiz.izquierdo = insertar(raiz.izquierdo, valor)
    else:
        raiz.derecho = insertar(raiz.derecho, valor)
    return raiz
```

```python
Datos = ["*", "+", "-", "7", "3", "5", "2"]
raiz = None
for i in Datos:
    raiz = insertar(raiz, i)
```

```python
raiz = nodo("*")

raiz.izquierdo = nodo("+")
raiz.derecho = nodo("-")
raiz.izquierdo.izquierdo = nodo("7")
raiz.izquierdo.derecho = nodo("3")

raiz.derecho.izquierdo = nodo("5")
raiz.derecho.derecho = nodo("2")
```

```python
print("\nRecorrer en PreOrden")


def PreOrden(raiz):
    if raiz:
        print(raiz.valor, end=' ')
        PreOrden(raiz.izquierdo)
        PreOrden(raiz.derecho)


PreOrden(raiz)

print("\nRecorrer en InOrden")


def inO(raiz):
    if raiz:
        yield "("
        yield from inO(raiz.izquierdo)
        yield raiz.valor
        yield from inO(raiz.derecho)
        yield ")"
def InOrden(raiz):
    return "".join(inO(raiz))

print(InOrden(raiz))

print("\nRecorrer en PostOrden")
def PostOrden(raiz):
    if raiz:
        PostOrden(raiz.izquierdo)
        PostOrden(raiz.derecho)
        print(raiz.valor, end=' ')


PostOrden(raiz)
```

```python
print("Evaluar expresion")

print(f"\nExpresion: {InOrden(raiz)}")

print(f"\nResultado {eval(InOrden(raiz))}")
```
